# 机器学习实验：决策树
### 利用 CART（分类和回归树）算法对鸢尾花数据集进行分类（调用sklearn库实现）

```python
'''
    注意：
    （1）Windows下，graphviz库需要在高级系统设置中手动添加进环境变量才能正常使用
    （2）代码参考书籍《Hands-On Machine Learning with Scikit-Learn, Keras, and TensorFlow》
'''
```

```python
# 导库
from graphviz import Source
from sklearn.datasets import load_iris
from sklearn.tree import export_graphviz
from sklearn.tree import DecisionTreeClassifier
```

```python
# 导入鸢尾花数据集
iris = load_iris()
```

```python
# 数据预处理（只用其中两个属性作为输入来分类）
x = iris.data[:, 2:]
y = iris.target
```

```python
# 构建决策树
tree_clf = DecisionTreeClassifier(max_depth=2)
tree_clf.fit(x, y)
```

```python
export_graphviz(tree_clf,
                out_file="dataset/iris_tree.dot",
                feature_names=["petal length (cm)", "petal width (cm)"],
                class_names=iris.target_names,
                rounded=True,
                filled=True
                )
```
